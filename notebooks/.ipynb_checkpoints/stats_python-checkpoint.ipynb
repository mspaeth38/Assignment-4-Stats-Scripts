# 3.1.2. Hypothesis testing: comparing two groups

```python
# importing stats from scipy to do the statistical tests
from scipy import stats
```

## 3.1.2.1. Student’s t-test: the simplest statistical test 

```python
# is the mean of the population different than zero?
stats.ttest_1samp(data['VIQ'], 0)   
```

```python
# Yes, the mean value is statistically different than zero, because the p-value is less than 0.05
```

### 2-sample t-test: testing for difference across populations

```python
# seeing if the mean VIQ in the male and female populations are different from each other
female_viq = data[data['Gender'] == 'Female']['VIQ']
male_viq = data[data['Gender'] == 'Male']['VIQ']
stats.ttest_ind(female_viq, male_viq)
```

```python
# the p-value is 0.045 which is less than 0.05 so the mean values of the male and female populations
# for VIQ are statistically different from each other
```

## 3.1.2.2. Paired tests: repeated measurements on the same individuals

```python
# This tests if FSIQ and PIQ are different from each other (2-sample t-test)
stats.ttest_ind(data['FSIQ'], data['PIQ'])  
```

```python
# However this isn't the best test to use because it neglects the fact that there are links 
# between the observations (from the same individual)

#Therefore, we need to use a paired t-test
stats.ttest_rel(data['FSIQ'], data['PIQ'])
```

```python
# the results show there is a difference since p-value is 0.082
```

```python
#the paired t-test is equivalent to a 1-sample test on the difference
stats.ttest_1samp(data['FSIQ'] - data['PIQ'], 0)  
```

```python
# It is important to note that t-tests assume Gaussian errors and so we can use a 
# Wilcox signed-rank test to remove that assumption
stats.wilcoxon(data['FSIQ'], data['PIQ'])   
```

## 3.1.3.1. “formulas” to specify statistical models in Python

### A simple linear regression

```python
# Testing if y is a linear function of x
# use stats module to fit a linear model and also test that the coefficient in y=mx+b is non-zero.
import numpy as np
x = np.linspace(-5, 5, 20)
np.random.seed(1)
# normal distributed noise
y = -5 + 3*x + 4 * np.random.normal(size=x.shape)
# Create a data frame containing all the relevant variables
data = pandas.DataFrame({'x': x, 'y': y})
```

```python
# using stats models (had to download the package)
import statsmodels.formula.api as sta
from statsmodels.formula.api import ols
model = ols("y ~ x", data).fit()
```

```python
# shows stats from the linear fit we gave the data
print(model.summary())
```

## Categorical variables: comparing groups or multiple categories

```python
# We are looking at the brain_size.csv file again and plotting a linear regression from the data
data = pandas.read_csv('brain_size.csv', sep=';', na_values=".")
model = ols("VIQ ~ Gender + 1", data).fit()
print(model.summary())
```

```python
# to compare between IQ's we would need a long form table: each type of IQ needs to be identifiable
# by a category
data_fisq = pandas.DataFrame({'iq': data['FSIQ'], 'type': 'fsiq'})
data_piq = pandas.DataFrame({'iq': data['PIQ'], 'type': 'piq'})
data_long = pandas.concat((data_fisq, data_piq))
print(data_long)  
```

```python
#then we can use this data_long to model the linear regression
model = ols("iq ~ type", data_long).fit()
print(model.summary()) 
```

## 3.1.3.2 Multiple Regression

```python
# reading the iris.csv file
data = pandas.read_csv('iris.csv')
model = ols('sepal_width ~ name + petal_length', data).fit()
print(model.summary()) 
```

## 3.1.3.3. Post-hoc hypothesis testing: analysis of variance (ANOVA)

```python
print(model.f_test([0, 1, -1, 0])) 
```

```python
# the difference is not significant because p is greater than 0.05
```

## 3.1.4. More visualization: seaborn for statistical exploration

```python
import urllib.request
import os
import matplotlib.pyplot as plt
```

```python
if not os.path.exists('wages.txt'):
    #download the file 
    urllib.request.urlretrieve('http://lib.stat.cmu.edu/datasets/CPS_85_Wages',
                       'wages.txt')
```

```python
names = [
    'EDUCATION: Number of years of education',
    'SOUTH: 1=Person lives in South, 0=Person lives elsewhere',
    'SEX: 1=Female, 0=Male',
    'EXPERIENCE: Number of years of work experience',
    'UNION: 1=Union member, 0=Not union member',
    'WAGE: Wage (dollars per hour)',
    'AGE: years',
    'RACE: 1=Other, 2=Hispanic, 3=White',
    'OCCUPATION: 1=Management, 2=Sales, 3=Clerical, 4=Service, 5=Professional, 6=Other',
    'SECTOR: 0=Other, 1=Manufacturing, 2=Construction',
    'MARR: 0=Unmarried,  1=Married',
]

short_names = [n.split(':')[0] for n in names]

data = pandas.read_csv('wages.txt', skiprows=27, skipfooter=6, sep=None,
                       header=None)
data.columns = short_names

# Log-transform the wages, because they typically are increased with
# multiplicative factors
import numpy as np
data['WAGE'] = np.log10(data['WAGE'])
```

```python
import seaborn
```

```python
seaborn.lmplot(y='WAGE', x='EDUCATION', data=data)
plt.show()
```

```python
import seaborn
seaborn.pairplot(data, vars=['WAGE', 'AGE', 'EDUCATION'],
                 kind='reg')
```

```python
seaborn.pairplot(data, vars=['WAGE', 'AGE', 'EDUCATION'],
                 kind='reg', hue='SEX')
```

## 3.1.4.2. lmplot: plotting a univariate regression

```python
seaborn.lmplot(y='WAGE', x='EDUCATION', data=data)
```
